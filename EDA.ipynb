```python
def extract_key_phrases(texts):
    patterns = {
        'bed_type': r'\b(?:KG|KING|QN|QUEEN|DBL|DOUBLE|TWN|TWIN|SGL|SINGLE)\b',
        'room_type': r'\b(?:STE|SUITE|RM|ROOM|EXEC|EXECUTIVE)\b',
        'amenities': r'\b(?:BALCONY|WIFI|HDTV|WET-BAR)\b',
        'accessibility': r'\b(?:ACCESS|MOB|HEAR)\b',
        'size': r'\b\d+\s*(?:SQ\s*FT|SQFT)\b'
    }
    
    results = []
    for text in texts:
        matches = {}
        for attr_type, pattern in patterns.items():
            matches[attr_type] = re.findall(pattern, text.upper())
        results.append(matches)
    
    return pd.DataFrame(results)
```

```python
#%%
def create_pattern_from_list(attribute_list):
    """Convert list of attributes into regex patterns including common abbreviations"""
    patterns = set()
    
    for attr in attribute_list:
        if pd.isna(attr):
            continue
            
        # Convert to string and uppercase
        attr = str(attr).upper()
        
        # Split camelCase
        words = re.findall('[A-Z][^A-Z]*', attr)
        if not words:
            words = [attr]
        
        # Add full term
        patterns.add(''.join(words))
        
        # Add common abbreviations
        if len(words) > 1:
            abbrev = ''.join(word[0] for word in words)
            if len(abbrev) > 1:
                patterns.add(abbrev)
        
        # Add common shortenings
        for word in words:
            if len(word) > 3:
                patterns.add(word[:3])
                
        # Add the full words
        patterns.update(words)
    
    return r'\b(?:' + '|'.join(patterns) + r')\b'
```

```python

def extract_key_phrases(texts, room_types_list, bed_types_list): # amenities_list):
    
    patterns = {
        'bed_type': create_pattern_from_list(bed_types_list),
        'room_type': create_pattern_from_list(room_types_list)#,
        #'amenities': create_pattern_from_list(amenities_list),
        #'accessibility': r'\b(?:ACCESS|MOB|HEAR)\b',
        #'size': r'\b\d+\s*(?:SQ\s*FT|SQFT)\b'
    }
    
    results = []
    for text in texts:
        matches = {}
        for attr_type, pattern in patterns.items():
            matches[attr_type] = re.findall(pattern, text.upper())
        results.append(matches)
    
    return pd.DataFrame(results)
```

```python
# Sample texts with their labels and calculate accuracy
def evaluate_extraction_accuracy(dataset, sample_size=100):
    # Sample from dataset
    sampled_data = dataset.groupby('Guest Room Info').apply(
        lambda x: pd.DataFrame({
            'Room Description': x['Room Description'].sample(min(3, len(x))),
            'Guest Room Info': x.name
        })
    ).reset_index(drop=True)
    
    # Extract attributes
    extracted_results = extract_key_phrases(
        sampled_data['Room Description'].tolist(), 
        room_types_list, 
        bed_types_list
    )
    
    # Calculate accuracy
    correct = 0
    total = len(sampled_data)
    
    for i, row in sampled_data.iterrows():
        true_label = row['Guest Room Info']
        predicted_types = extracted_results.iloc[i]['room_type']
        
        # Convert to uppercase for comparison
        if any(true_label.upper() in pred.upper() for pred in predicted_types):
            correct += 1
    
    accuracy = correct / total
    print(f"\nAccuracy: {accuracy:.2%}")
    
    # Print some examples
    print("\nSample Predictions:")
    for i in range(min(5, len(sampled_data))):
        print(f"\nDescription: {sampled_data.iloc[i]['Room Description']}")
        print(f"True Label: {sampled_data.iloc[i]['Guest Room Info']}")
        print(f"Predicted: {extracted_results.iloc[i]['room_type']}")
    
    return accuracy, sampled_data, extracted_results
```

def preprocess_pattern_list(pattern_list):
    """
    Convert camelCase/PascalCase patterns to lowercase and split into individual words
    Example: 'KingBed' -> ['king', 'bed']
    """
    processed = []
    for pattern in pattern_list:
        # Handle NaN values and empty strings
        if pd.isna(pattern) or str(pattern).strip() == '':
            continue
            
        # Convert to string and strip whitespace
        pattern = str(pattern).strip()
        
        # Split camelCase/PascalCase
        words = re.findall('[A-Z][^A-Z]*', pattern)
        
        # Only add non-empty patterns
        if words:  # If words were found by the regex
            processed_pattern = ' '.join(word.lower() for word in words).strip()
            if processed_pattern:  # If the result isn't empty
                processed.append(processed_pattern)
        else:  # If no camelCase/PascalCase found, just lowercase the whole thing
            if pattern:  # If pattern isn't empty
                processed.append(pattern.lower().strip())
    
    # Remove duplicates and empty strings
    return [p for p in list(set(processed)) if p.strip()]

```python
# Function to clean and tokenize text
def clean_text(texts):
    cleaned_texts = []
    for text in texts:
        if pd.isna(text):  # Handle NaN values
            cleaned_texts.append("")
            continue
            
        # Convert to lowercase
        text = text.lower()
        
        # Replace specific compound terms
        compound_terms = {
            'bed room': 'bedroom'
        }
        
        for term, replacement in compound_terms.items():
            text = text.replace(term, replacement)
            
        cleaned_texts.append(text)
    
    return cleaned_texts
```

```python
# SPaCy - Extract attributes using spaCy
def analyze_with_spacy(texts):
    # Load transformer-based English model
    try:
        nlp = spacy.load("en_core_web_trf")
    except OSError:
        print("Downloading en_core_web_trf model...")
        spacy.cli.download("en_core_web_trf")
        nlp = spacy.load("en_core_web_trf")
    
    # Preprocess the pattern lists
    processed_room_types = [room_type.lower() for room_type in room_types_list]
    processed_bed_types = [bed_type.lower() for bed_type in bed_types_list]
    #processed_room_views = room_view_list
    #processed_amenities = room_amenities_list
    
    # Custom labels for hospitality domain
    room_patterns = [
        {"label": "ROOM_TYPE", "pattern": [{"LOWER": {"IN": processed_room_types}}]},
        {"label": "BED_TYPE", "pattern": [{"LOWER": {"IN": processed_bed_types}}]}#,
        #{"label": "VIEW_TYPE", "pattern": [
        #    {"LOWER": {"IN": processed_room_views}},
        #    {"LOWER": "view", "OP": "?"}
        #]},
        #{"label": "AMENITY", "pattern": [{"LOWER": {"IN": processed_amenities}}]}
    ]
    
    # Add patterns to pipeline
    if "entity_ruler" not in nlp.pipe_names:
        ruler = nlp.add_pipe("entity_ruler", before="ner")
    else:
        ruler = nlp.get_pipe("entity_ruler")
    ruler.add_patterns(room_patterns)
    
    results = defaultdict(list)
    
    # Process each text with batch processing for better performance
    batch_size = 32  # Adjusting a random number based on potential available RAM
    for i in tqdm(range(0, len(texts), batch_size), desc="Processing with spaCy"):
        batch_texts = texts[i:i + batch_size]
        batch_texts = [text if pd.notna(text) else "" for text in batch_texts]
        
        # Process batch
        docs = nlp.pipe(batch_texts)
        
        for doc in docs:
            entities = {
                "text": doc.text,
                "room_types": [],
                "bed_types": [],
                #"view_types": [],
                #"amenities": [],
                "other_entities": []
            }
            
            for ent in doc.ents:
                if ent.label_ == "ROOM_TYPE":
                    entities["room_types"].append(ent.text.lower())
                elif ent.label_ == "BED_TYPE":
                    entities["bed_types"].append(ent.text.lower())
                #elif ent.label_ == "VIEW_TYPE":
                #    entities["view_types"].append(ent.text.lower())
                #elif ent.label_ == "AMENITY":
                #    entities["amenities"].append(ent.text.lower())
                else:
                    entities["other_entities"].append((ent.text.lower(), ent.label_))
            
            for key, value in entities.items():
                results[key].append(value)
    
    return pd.DataFrame(results)

```

```python
import spacy
from tqdm import tqdm
from collections import defaultdict

def clean_text(texts):
    """Clean and standardize text descriptions"""
    cleaned_texts = []
    for text in texts:
        if pd.isna(text):
            cleaned_texts.append("")
            continue
            
        # Convert to lowercase
        text = text.lower()
        
        # Replace specific compound terms
        compound_terms = {
            'bed room': 'bedroom'
        }
        
        for term, replacement in compound_terms.items():
            text = text.replace(term, replacement)
            
        cleaned_texts.append(text)
    
    return cleaned_texts

def setup_ner_patterns(nlp, attribute_lists):
    """Setup NER patterns for all attribute types"""
    patterns = []
    
    for attr_type, attr_list in attribute_lists.items():
        for attr in attr_list:
            if pd.notna(attr):
                words = re.findall('[A-Z][^A-Z]*', str(attr))
                full_form = ' '.join(words) if words else str(attr)
                
                patterns.append({
                    "label": attr_type.upper(),
                    "pattern": full_form,
                    "id": str(attr)
                })
                
                if ' ' in full_form:
                    patterns.append({
                        "label": attr_type.upper(),
                        "pattern": full_form.split(),
                        "id": str(attr)
                    })
    
    if "entity_ruler" not in nlp.pipe_names:
        ruler = nlp.add_pipe("entity_ruler", before="ner")
    else:
        ruler = nlp.get_pipe("entity_ruler")
    ruler.add_patterns(patterns)
    return nlp

def extract_attributes_ner(texts, attribute_lists):
    """Extract attributes using spaCy NER with custom patterns"""
    try:
        nlp = spacy.load("en_core_web_trf")
    except OSError:
        print("Downloading en_core_web_trf model...")
        spacy.cli.download("en_core_web_trf")
        nlp = spacy.load("en_core_web_trf")
    
    # Clean texts first
    cleaned_texts = clean_text(texts)
    
    # Setup patterns
    nlp = setup_ner_patterns(nlp, {
        'room_type': room_types_list,
        'bed_type': bed_types_list,
        'amenity': room_amenities_list
    })
    
    results = []
    
    # Process in batches
    batch_size = 32
    for i in tqdm(range(0, len(cleaned_texts), batch_size), desc="Processing with NER"):
        batch_texts = cleaned_texts[i:i + batch_size]
        
        docs = nlp.pipe(batch_texts)
        
        for doc in docs:
            entities = {
                "text": doc.text,
                "room_types": [],
                "bed_types": [],
                "amenities": []
            }
            
            for ent in doc.ents:
                if ent.label_ == "ROOM_TYPE":
                    entities["room_types"].append({
                        'text': ent.text,
                        'normalized': ent.ent_id_
                    })
                elif ent.label_ == "BED_TYPE":
                    entities["bed_types"].append({
                        'text': ent.text,
                        'normalized': ent.ent_id_
                    })
                elif ent.label_ == "AMENITY":
                    entities["amenities"].append({
                        'text': ent.text,
                        'normalized': ent.ent_id_
                    })
            
            results.append(entities)
    
    return pd.DataFrame(results)

```

```python
# NER Spacy with confidence scores and rule-based matches
def extract_attributes_ner(texts, attribute_lists):
    """Extract attributes using spaCy NER with confidence scores"""
    try:
        nlp = spacy.load("en_core_web_trf")
    except OSError:
        print("Downloading en_core_web_trf model...")
        spacy.cli.download("en_core_web_trf")
        nlp = spacy.load("en_core_web_trf")
    
    # Clean texts first
    cleaned_texts = clean_text(texts)
    
    # Setup patterns
    patterns = []
    for attr_type, attr_list in attribute_lists.items():
        for attr in attr_list:
            if pd.notna(attr):
                attr_text = str(attr).lower()
                patterns.append({
                    "label": attr_type.upper(),
                    "pattern": [{"LOWER": attr_text}],
                    "id": str(attr),
                    "source": "rule"  # Mark as rule-based match
                })
    
    if "entity_ruler" not in nlp.pipe_names:
        ruler = nlp.add_pipe("entity_ruler", before="ner")
    else:
        ruler = nlp.get_pipe("entity_ruler")
    ruler.add_patterns(patterns)
    
    results = []
    batch_size = 32
    
    for i in tqdm(range(0, len(cleaned_texts), batch_size), desc="Processing with NER"):
        batch_texts = cleaned_texts[i:i + batch_size]
        docs = nlp.pipe(batch_texts)
        
        for doc in docs:
            entities = {
                "text": doc.text,
                "room_types": [],
                "bed_types": [],
                "amenities": []
            }
            
            # Check for accessibility keywords
            is_accessible = 'access' in doc.text.lower() or 'ada' in doc.text.lower()
            
            for ent in doc.ents:
                # Default confidence for rule-based matches
                score = 0.8 if getattr(ent, 'source', None) == 'rule' else 1.0
                
                # Try to get neural confidence if available
                try:
                    if hasattr(ent, 'tensor') and ent.tensor is not None:
                        score = float(ent.tensor.max())
                except (ValueError, AttributeError):
                    pass
                
                if ent.label_ == "ROOM_TYPE":
                    entities["room_types"].append({
                        'text': ent.text.lower(),
                        'score': score,
                        'source': getattr(ent, 'source', 'neural')
                    })
                elif ent.label_ == "BED_TYPE":
                    entities["bed_types"].append({
                        'text': ent.text.lower(),
                        'score': score,
                        'source': getattr(ent, 'source', 'neural')
                    })
            
            # Add Accessible Room if keywords found
            if is_accessible:
                entities["room_types"].append({
                    'text': 'accessible room',
                    'score': 1.1,
                    'source': 'rule'
                })
            
            # Sort by confidence score
            for key in entities.keys():
                if isinstance(entities[key], list):
                    entities[key] = sorted(entities[key], key=lambda x: x['score'], reverse=True)
            
            results.append(entities)
    
    return pd.DataFrame(results)
```

```python

def evaluate_ner_accuracy(ner_results, full_data):
    """Evaluate NER accuracy with confidence scores"""
    if len(ner_results) != len(full_data):
        print(f"Warning: Results ({len(ner_results)}) and samples ({len(full_data)}) length mismatch")
        n_samples = min(len(ner_results), len(full_data))
        ner_results = ner_results.iloc[:n_samples]
        full_data = full_data.iloc[:n_samples]
    
    correct_any = 0
    correct_top = 0
    total = len(full_data)
    
    print("\nNER Extraction Evaluation:")
    print("==========================")
    
    for i, (_, row) in enumerate(full_data.iterrows()):
        true_label = str(row['Guest Room Info']).lower()
        predictions = ner_results.iloc[i]['room_types']
        
        if not predictions:
            continue
        
        # Check top prediction
        top_pred = predictions[0]['text'].lower()
        top_match = true_label in top_pred or top_pred in true_label
        
        # Check any prediction
        any_match = any(true_label in pred['text'].lower() or pred['text'].lower() in true_label 
                       for pred in predictions)
        
        correct_top += top_match
        correct_any += any_match
        
        if i < 5:
            print(f"\nDescription: {row['Room Description']}")
            print(f"True Label: {true_label}")
            print(f"Top Prediction: {top_pred} (score: {predictions[0]['score']:.3f}, source: {predictions[0]['source']})")
            print("All Predictions:", [(p['text'], f"{p['score']:.3f}", p['source']) for p in predictions])
            print(f"Top Match: {'âœ“' if top_match else 'âœ—'}")
            print(f"Any Match: {'âœ“' if any_match else 'âœ—'}")
    
    top_accuracy = correct_top / total
    any_accuracy = correct_any / total
    
    print(f"\nTop Prediction Accuracy: {top_accuracy:.2%}")
    print(f"Any Match Accuracy: {any_accuracy:.2%}")
    
    return {"top_accuracy": top_accuracy, "any_accuracy": any_accuracy}
```
